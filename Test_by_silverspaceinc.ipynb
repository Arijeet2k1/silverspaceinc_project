```python
#Converting the tsv file into pandas dataframe
import pandas as pd
df = pd.read_csv("C:\\Users\\User\\Downloads\\correct_twitter_201904.tsv", sep='\t')
```

```python
# Step 1: Input the term you're searching for
term = input("Enter the term to search for: ")

# Step 2: Filtered rows where the 'text' column contains the specific term
filtered_df = df[df['text'].str.contains(term, case=False, na=False)].copy()

# --- Answering the Questions ---

# Question 1: How many tweets were posted containing the term on each day?
filtered_df.loc[:, 'date'] = pd.to_datetime(filtered_df['ts1']).dt.date
tweets_per_day = filtered_df.groupby('date')['id'].count().reset_index()
tweets_per_day.columns = ['date', 'tweet_count']
print("\nTweets containing the term on each day:")
print(tweets_per_day)

# Question 2: How many unique users posted a tweet containing the term?
unique_users = filtered_df['author_id'].nunique()
print("\nNumber of unique users who posted a tweet containing the term:", unique_users)

# Question 3: How many likes did tweets containing the term get, on average?
average_likes = filtered_df['like_count'].mean()
print("\nAverage likes for tweets containing the term:", round(average_likes, 2))

# Question 4: Where (in terms of place IDs) did the tweets come from?
tweets_by_place = filtered_df.groupby('place_id')['id'].count().reset_index()
tweets_by_place.columns = ['place_id', 'tweet_count']
print("\nTweets by place (in terms of place IDs):")
print(tweets_by_place)

# Question 5: What times of day were the tweets posted at?
filtered_df['created_at'] = pd.to_datetime(filtered_df['created_at'], errors='coerce')
filtered_df['hour'] = filtered_df['created_at'].dt.hour
tweets_by_hour = filtered_df.groupby('hour')['id'].count().reset_index()
tweets_by_hour.columns = ['hour', 'tweet_count']
print("\nTweets by time of day (hour):")
print(tweets_by_hour)

# Question 6: Which user posted the most tweets containing the term?
tweets_by_user = filtered_df.groupby('author_id')['id'].count().reset_index()
tweets_by_user.columns = ['author_id', 'tweet_count']
most_active_user = tweets_by_user.loc[tweets_by_user['tweet_count'].idxmax()]
print(f"\nUser with ID {most_active_user['author_id']} posted the most tweets containing the term, with {most_active_user['tweet_count']} tweets.")
```
